```python
reorder = int(input("Enter reorder point "))
print("You entered ", reorder)
```

```python
def demand_funct():
    a= random.randint(1,100)
    if a >= 92:
        demand = 4
    elif a >= 71:
        demand = 3
    elif a >= 36:
        demand = 2
    elif a >= 11:
        demand = 1
    else:
        demand = 0
    return demand

def leadtime_funct():
    b= random.randint(1,100)
    if b >= 91:
        leadtime = 3
    elif b >= 61:
        leadtime = 2
    else:
        leadtime = 1
    return leadtime
```

```python
import random
receiving= False
shortage= 0
acc_inventory=0
acc_shortage=0
for days in range(25):
    demand= demand_funct()
    print ('Day:', days+1,' Current Shortage: ', shortage)
    if receiving == True and leadtime == 0:
        inventory += level
        print('A new restocking order has been received, new inventory:', inventory)
        receiving= False
    elif receiving == True and leadtime != 0:
        print('A new restocking is expected in', leadtime, 'days.')
        leadtime -=1
    elif receiving == False:
        print('No restocking is expected for today.')
    if shortage > 0 and inventory >= shortage:
        inventory-= shortage
        print ('The shortage of', shortage, 'has been delivered now')
        shortage=0
    print('Begining Inventory is:', inventory)
    print('Daily Demand: ', demand)
    if inventory >= demand:
        inventory -= demand 
        print('Todays order of', demand, 'items has been fulfilled.', 'Ending Inventory=', inventory )
    else:
        print(inventory, ' items have been delivered')
        shortage+= (demand - inventory)
        inventory=0
        print('Todays order of', demand, 'items cant be completely fulfilled because of low stock.', 'Shortage =', shortage, 'Ending Inventory=', inventory )
    if inventory<=reorder and receiving == False:
            receiving = True
            leadtime= leadtime_funct()
            print('Current stock is equal to or below reorder point, re-order of: ', level, 'items will be fulfilled in: ', leadtime, 'days.')
    print ('End of day',days+1)
    acc_inventory += inventory
    acc_shortage += shortage
print('simulation completed', 'Average Ending Inventory= ', acc_inventory/25, 'Average Shortage= ', acc_shortage/25)    
    
```
